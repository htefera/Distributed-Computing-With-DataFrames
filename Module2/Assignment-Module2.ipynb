### Question 3

**Which "Unit Type" is the most common?**<br>
**Answer == Engine**

```
%sql
-- ANSWER
SELECT `Unit Type`, COUNT(*) AS TOTAL_UNIT_TYPE FROM FIRECALLS GROUP BY `UNIT TYPE` ORDER BY TOTAL_UNIT_TYPE  DESC
```
