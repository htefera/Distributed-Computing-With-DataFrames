## ![Spark Logo Tiny](https://files.training.databricks.com/images/105/logo_spark_tiny.png) Getting Data into Databricks

There are a number of ways of getting data into Databricks.  In this lesson, you'll upload a flat CSV file into Databricks before creating a table and querying it.

## Download Dataset

Download a subset of the Data SF's Fire Department Calls for Service [here](https://s3-us-west-2.amazonaws.com/davis-dsv1071/data/fire-calls/fire-calls-truncated-comma-no-spaces.csv). This dataset is about 85 MB.

The entire dataset is ~1.7GB, and can be found on [Data SF](https://data.sfgov.org/Public-Safety/Fire-Department-Calls-for-Service/nuek-vuh3/data)

## ![Spark Logo Tiny](https://files.training.databricks.com/images/105/logo_spark_tiny.png) Querying Data

Now you can access the file you uploaded.

Take a look at the first few lines of the data.

```
%sql
SELECT * FROM firecallsuploaded LIMIT 10
```

Look at the number of emergency calls by call type.

```
%sql
SELECT `Call_Type`, count(1) as total_emergency_calls_made 
FROM firecallsuploaded 
GROUP BY `Call_Type` 
ORDER BY total_emergency_calls_made DESC LIMIT 10
```

Now look at calls by neighborhood.

```
%sql
SELECT `Neighborhooods_-_Analysis_Boundaries` as neighborhood, count(1) as total_emergency_calls_made 
FROM firecallsuploaded 
GROUP BY `Neighborhooods_-_Analysis_Boundaries` 
ORDER BY total_emergency_calls_made DESC LIMIT 10
```
